# Calculate summary statistics (mean, median, mode, standard deviation)       for a dataset

```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('test.csv')
```

```python
df
```

```python
# Calculate summary statistics for each variable
for col in df.columns:
    summary = df[col].describe()
    print(col, summary)
```

```python
# Calculate summary statistics for multiple variables
df_summary = df[['Age', 'Pclass', 'PassengerId','Ticket','Fare']].describe()
print(df_summary)
```
