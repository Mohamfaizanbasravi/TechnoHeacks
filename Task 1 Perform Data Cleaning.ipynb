# Clean a dataset by removing missing values and outliers.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('test.csv')
```

```python
df
```

```python
# Identify the missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Remove the rows with missing values
df = df.dropna()
```

```python
# Print the number of missing values in each column
print(missing_values.sum(axis=0))
```

```python
# Identify outliers using z-score
z_score = (df - df.mean()) / df.std()
outliers = df[z_score.abs() > 3]
```

```python
# Print the number of outliers in each column
print(outliers.shape[0])
```

```python
# Impute the outliers with the mean value of the column
df.update(df.fillna(df.mean()))
```

```python
df
```

```python
# Save the {cleaned dataset} to a new CSV file 
df.to_csv('clean_titanic_dataset.csv', index=False)
```
