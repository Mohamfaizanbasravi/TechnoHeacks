# Create a histogram or bar chart to visualizethe distribution of data in a dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
df = pd.read_csv('iris.csv')
```

```python
df
```

```python
# Create a histogram of the SepalLengthCm column
plt.hist(df['SepalLengthCm'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of sepal length in the Iris dataset')
plt.show()

```

```python
# Create a bar chart of the Species column
plt.bar(df['Species'].value_counts().index, df['Species'].value_counts().values)
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Bar chart of species in the Iris dataset')
plt.show()
```

```python
# Get the names of all numeric columns
numeric_columns = df.select_dtypes(include='number').columns
```

```python
# Create a new DataFrame with only the numeric columns
df_numeric = df[numeric_columns]
```

```python
# Create a histogram for each numeric column
for column in df_numeric.columns:
    plt.hist(df_numeric[column])
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column}')
    plt.show()
```

```python
# Create a bar chart for the Species column
plt.bar(df['Species'].value_counts().index, df['Species'].value_counts().values)
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Bar chart of species in the Iris dataset')
plt.show()
```
